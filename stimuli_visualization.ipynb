```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import sys
from image_tools.loading import load_nsd_images, load_majaj_images

ROOT = os.getenv('BONNER_ROOT_PATH')
sys.path.append(ROOT)


def plot_images(image_paths, name, rows=4, cols=5):
    """
    Plots a grid of images specified by the paths in image_paths.

    :param image_paths: List of paths to the images.
    :param rows: Number of rows in the grid.
    :param cols: Number of columns in the grid.
    """
    fig, axes = plt.subplots(rows, cols, figsize=(15, 10))
    axes = axes.ravel()

    for i, img_path in enumerate(image_paths):
        img = mpimg.imread(img_path)
        axes[i].imshow(img)
        axes[i].axis('off')

    plt.tight_layout()
    plt.savefig(name, dpi=200) 
    plt.show()

```

```python
# Example usage
image_paths = load_nsd_images()[:20] # Replace with actual paths
plot_images(image_paths, name='nsd_images.png')
```

```python
# Example usage
import random
all_image_paths = load_majaj_images() # Replace with actual paths
image_paths = random.sample(all_image_paths, 20)
plot_images(image_paths, name='majaj_images.png')
```
